```python
import numpy as np
import time
```

```python
def timeNorm(x):
    '''
    Returns the time to generate
    a random normal vector of size x
    '''
    t0 = time.time()
    np.random.normal(0,1,(x,))
    return time.time() - t0
```

```python
xvals = [x for x in range(1,10001)]
yvals = [timeNorm(x) for x in xvals]
```

```python
import matplotlib.pyplot as pplot
pplot.plot(xvals,yvals)
pplot.show()
```
