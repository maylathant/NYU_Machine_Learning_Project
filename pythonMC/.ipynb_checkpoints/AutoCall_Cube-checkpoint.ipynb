```python
import FinObj as fin
import numpy as np
import riskEngine as risk
import smoothing as smooth
```

```python
#Map for units of time
t_map = {'2y':0.5,'year':1,'semester':2,'quarter':4,'month':12,'week':52,'2day':180,'day':360}
```

```python
#MC Parameters
paths = 5000
t_del = 1

t_unit = 12 #Unit of time: default = 1 for years, 12 for months, 360 for days etc...
call_dates = 4 #Number of call dates per unit time
time_s = 12 #Total number of time steps in t_units
```

```python
#Stock Parameters
spot = 120
vol = 0.2 #vol per year
rate = 0 #Rate in years

#Share object
share1 = fin.Stock(spot,vol,t_unit)
```

```python
#Autocall Parameters
C = 0.25 #Annual rate coupon
ssize = 1 #Notional
downin = 0.7 #Put down and in
mature = time_s #Maturity (initially assume last day of time step)
strike = 100 #Strike price


#autocall object
autoc = fin.AutoCall(share1, C, fin.schFixed, ssize, downin, mature, strike, call_dates, t_unit)
```

```python
#Define spot and vol for cube
maxi, mini = 125, 50 #Max and min values
resolution = 25 #Number of grid points
spots = np.linspace(mini,maxi,resolution)

#Define vols for cube
v_maxi, v_mini = 0.6, 0.1 #Max and min values
v_resolution = 5 #Number of grid points
vols = np.linspace(v_mini,v_maxi,v_resolution)
```

```python
sv_cube=risk.getSpotVol(share1,autoc,paths,time_s,t_del,spots,vols,ini_t=8)
```

```python
#Plot varying spot
import matplotlib.pyplot as pplot
#Plot each valuation
for v in range(0,len(vols)):
    pplot.plot(spots,sv_cube[v],label='Vol = '+str(vols[v]))
pplot.legend()
pplot.show()
```

```python
#Plot varying vol
iterator = [x for x in range(0,len(spots))][0::5]
for v in iterator:
    pplot.plot(vols,sv_cube[:,v],label='Spot = '+str(spots[v]))
pplot.legend()
pplot.show()
```
